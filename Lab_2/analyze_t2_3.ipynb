```python
import os
import re
import glob
import pandas as pd
import matplotlib.pyplot as plt

print(os.getcwd())

def get_info(file):
    prediction_rate=1   # for perfect condition
    speedup = 0
    cpi = 0
    lines = open(file).readlines()
    for line in lines:
        if line.find('bpred_dir_rate') > 0:
            prediction_rate = float(re.findall(r'0+\.\d+', line)[0])
            # print(prediction_rate)
        if line.find('sim_CPI') == 0:
            cpi = float(re.findall(r'\d+\.\d+', line)[0])
            # print(cpi)
    return prediction_rate,cpi

app_name_list=[
    "dijkstra_small",
    "stringsearch-cabce", 
    "qsort_small", 
    "gsm-untoast", 
    "jpeg-cjpeg"
    ]

ISSUE_WIDTH_list =[
  1,
  2,
  # 3,
  4,
  # 5,
#   6,
  # 7,
  8
]

DECODE_WIDTH_list =[
  1,
  2,
  # 3,
  4,
  # 5,
  # 6,
  # 7,
  8
]

COMMIT_WIDTH_list =[
  1,
  2,
  # 3,
  4,
  # 5,
  # 6,
  # 7,
  8
]


IF_Queue_size_list=[2,4,8]
RUU_size_list = [16,32,64,128]
ALUs_list = [1,2,3,4,5,6,7,8]
Mults_list = [1,2,3,4]
Issue_wrong_path_list = ["True", "False"]


flist = glob.glob("./simhome/configs/task_2_3/ifqs_2_ruus_16_lsq_8_alus_*_mults_*_iwp_False")
flist.sort(reverse=True)
print(len(flist))
# csv=open("task_2_3.csv",'w')
app_perf_map={}
app_perf_cpi_map={}
# print("app,"+",".join([p.split("/")[-1] for p in flist]), file=csv)

app = "dijkstra_small"
    # "dijkstra_small"
    # "stringsearch-cabce", 
    # "qsort_small", 
    # "gsm-untoast", 
    # "jpeg-cjpeg"
## to show the alus effection.

```

```python

def show_cpu(ifqs,ruus,lsq,mults,alu,iwp):
    cpi_list=[]
    for iss in ISSUE_WIDTH_list:
        item = "./simhome/configs/task_2_3/iss_%d_ifqs_%d_ruus_%d_lsq_%d_alus_%s_mults_%d_iwp_%s" \
            % (iss,ifqs,ruus,lsq,alu,mults,iwp)
        stat_file_path = os.path.join(item, "Stats_"+ app + ".txt")
        rate,cpi = get_info(stat_file_path)
        # print(rate, end=',', file=csv)
        cpi_list.append(cpi)
    print(cpi_list)
    return cpi_list

plt.figure(figsize=(12, 4))
plt.title(app)

plt.subplot(111)
print("ISSUE WIDTH vs CPI")
plt.ticklabel_format(useOffset=False, style='plain')
plt.title("ISSUE WIDTH vs ΔCPI,Decode=1,Commit=1,IFQS=8,RUU=128,LSQ=64,Mults=4,iALU=8,WRONG_PATH=False")
for app in app_name_list:
    y=show_cpu(8,128,64,4,8,"False")
    plt.plot(ISSUE_WIDTH_list, [ i - y[0] for i in y] , label=app)
plt.legend()
plt.xlabel("ISSUE_WIDTH")
plt.xticks(ISSUE_WIDTH_list)
plt.ylabel("ΔCPI")

plt.tight_layout() 
plt.show()

```
